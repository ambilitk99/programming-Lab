```python
# area of circle and triangle using module 
from Graphics.Circle1 import*
from Graphics.Rectangle import*
from Graphics.Graphics3d import sphere
from Graphics.Graphics3d import cuboid
while(1):
    print("\n 1.circle\n 2.rectangle\n 3.sphere\n 4.cuboid \n 5.exit")
    ch=int(input("Enter any operation :"))
    if ch==1:
        r=int(input("Enter any radius :"))
        carea=circlearea(r)
        cper= circleperimeter(r)
        print("Area of circle:",carea,"\n Perimeter of circle :",cper )
    elif ch==2:
        a=int(input("Enter length:"))
        b=int(input("Enter breadth:"))
        rarea=rectarea(a,b)
        rper=rectperimeter(a,b)
        print("Area of rectangle :",rarea,"\n perimeter of rectangle :",rper)
    elif ch==3:
        r=int(input("Enter any radius :"))
        sarea=sphere.spherearea(r)
        sper=sphere.sphereperimeter(r)
        print("Area of sphere:",sarea,"\n Perimeter ofsphere:",sper )
      
    elif ch==4:
        l=int(input("Enter length:"))
        b=int(input("Enter breadth:"))
        h=int(input("Enter height:"))
        cuarea=cuboid.cuboidarea(l,b,h)
        cuper=cuboid.cubperimeter(l,b,h)
        print("Area of cuboid:",cuarea,"\n perimeter of cuboid :",cuper)
    else:
        print("Exiting.....")
        break
    
        
```
